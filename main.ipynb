```python
#Модели
detector_craft = keras_ocr.detection.Detector()
detector_yolo = YOLO('models/yolo_best.pt')
recognizer_easyocr = easyocr.Reader(['en'])
```

```python
#Прогоняем по всем файлам в папке input_folder
input_folder = "dataset/"
output_folder = "output/"

file_names = os.listdir(input_folder)

files = os.listdir(input_folder)

yolo_predictions = []
#Прогон YOLO
for file in files:
    name, ext = os.path.splitext(file)
    #Детекция YOLO
    yolo_predict = detector_yolo(input_folder+file, save_crop=True)
    yolo_predictions.append(yolo_predict)
yolo_save_path = "runs/detect/predict/crops/full/"
```

```python
#Проверим какие найдены не были
for file in files:
    folder_path = yolo_save_path
    item_name = file
    if not os.path.exists(os.path.join(folder_path, item_name)):
        yolo_lost_list.append(file)

#Удалить при переносе в .py
yolo_lost_list
```

```python
craft_save_path_cropped = "output_craft_cropped/"
craft_save_path = "output_craft/"
```

```python
#Если есть те, где оно не нашло - запуск CRAFT и прогон им же
#Флаг запуска крафта
craft_flag = False
if yolo_lost_list != []:
    craft_flag = True
    

    for file in yolo_lost_list:
        cropped_img_list = []
        count = 0
        name, ext = os.path.splitext(file)

        image = keras_ocr.tools.read(f'input/{file}')
        boxes = detector_craft.detect(images=[image], detection_threshold=0.7, size_threshold=200)[0]
        canvas = keras_ocr.tools.drawBoxes(image, boxes, thickness=1)
        #Сохранение распознанного
        plt.imsave(f"output_craft/{name}.jpg", canvas)

        #Обрезка "квадратов"
        for cnt in boxes:
            padding = 5
            im2 = image.copy()
            x1 = int(cnt[0][0])-padding
            y1 = int(cnt[0][1])-padding
            x2 = int(cnt[2][0])+padding
            y2 = int(cnt[2][1])+padding
            w = x2-x1
            h = y2-y1
            cropped = im2[y1:y1 + h, x1:x1 + w]
            img = cropped
            cropped_img_list.append(img)
        #Сохраняем кропы
        for i in cropped_img_list:
            y,x,z = i.shape
            #Фильтр, отсеивающий часть кропов, которые явно не могут быть номером из-за размера изображения
            if x>z*1.5:
                #Отлавливаем ексепты, потому что некоторые кривые выделения крафта крашат сохранения
                try:
                    plt.imsave(f"output_craft_cropped/{name}-cropped-{count}.jpg", i)
                    count+=1
                except:
                    pass
else:
    print("Запуск CRAFTа пропущен")
```
